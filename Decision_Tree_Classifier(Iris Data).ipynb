**Importing the libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

**Reading the data**

```python
data = pd.read_csv("/content/Iris(Decision Tree).csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
#checking missing values
data.isnull().sum()
```

```python
x = data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
x[0:5]
```

```python
y = data['Species']
y[0:5]
```

**Training the data**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

**Modelling the data using Decision Tree Classifier**

```python
classifier = DecisionTreeClassifier(criterion = "entropy", max_depth = 4)
classifier.fit(x_train, y_train)
```

**Prediction & Evaluation**

```python
prediction = classifier.predict(x_test)
print("Accuracy:", accuracy_score(y_test,prediction))
```

**Visualization**

```python
#making an instance of the model
fn = data.columns[1:5]
cn = data["Species"].unique().tolist()
classifier.fit(x,y)
#setting dpi=300 to make image more clearer than default
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,10), dpi = 300)
tree.plot_tree(classifier, feature_names=fn, class_names=cn, filled = True);
```

**Checking for random prediction**

```python
x_new = [[5.7,4.0,1.6,0.3]]
new_pred = classifier.predict(x_new)
new_pred
```
